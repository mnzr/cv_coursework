## Task 1 [Done]
Plot the histogram of a range of images.

```python
# All imports
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
def rgb_histogram(filename):
    # Loading the image
    img=mpimg.imread(filename)    # numpy array

    red = img[:,:,0].flatten()
    green = img[:,:,1].flatten()
    blue = img[:,:,2].flatten()

    bins = np.linspace(0, 256, 500)    # start, end, sample size
    alpha = 0.3    # color opacity
    # Creating separate histograms for red, green, blue
    plt.hist(red, bins=bins, alpha=alpha, label='red', color='red')
    plt.hist(green, bins=bins, alpha=alpha, label='green', color='green')
    plt.hist(blue, bins=bins, alpha=alpha, label='blue', color='blue')
    # Plot config
    plt.legend(loc='upper right')
    plt.title("Histogram of {}".format(filename))
    plt.xlabel("Value In RGB Scale")
    plt.ylabel("Frequency")
    plt.show()
```

```python
rgb_histogram('Woman in Red and Yellow Saree Standing Smiling.jpeg')
rgb_histogram('Seashore Blue Body of Water during Daytime.jpeg')
rgb_histogram('Blue Body of Water Under Blue Sky.jpeg')
```
